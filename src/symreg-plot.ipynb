## Plot Results Obtained

```python
import random
import numpy as np
import copy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
problem = np.load('../data/problem_1.npz')
x = problem['x']
y = problem['y']
x.shape
```

## Plot Problem 1

```python
def plot_sine_comparison(x, y):
    plt.figure(figsize=(12, 6))
    
    # Plot original data
    plt.scatter(x[0], y, alpha=0.5, color='blue', label='Original data')
    
    x_continuous = np.linspace(x[0].min(), x[0].max(), 1000)
    y_sine = np.sin(x_continuous)
    
    # Plot sin function
    plt.plot(x_continuous, y_sine, 'r-', label='sin(x)', linewidth=2)
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Problem 1: comparison between original data and sin(x)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.axhline(y=1, color='gray', linestyle='--', alpha=0.3)
    plt.axhline(y=-1, color='gray', linestyle='--', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

plot_sine_comparison(x, y)
```
